```python
import pandas as pd
import numpy as np
```

```python
movies_df = pd.read_csv('animes.csv')
```

```python
movies_df.head()
```

```python
movies_df.info()
```

```python
movies_df.describe()
```

```python
movies_df = movies_df[['uid','title','synopsis','genre','episodes','score']]
movies_df = movies_df[movies_df['score']>=6.410000]
```

```python
movies_df.head()
```

```python
movies_df['genre'] = movies_df['genre'].apply(lambda x: x.replace("'","")).apply(lambda x: x.replace("[","")).apply(lambda x: x.replace("]",""))
movies_df.head()
```

```python
movies_df = movies_df.drop_duplicates()
movies_df.to_csv('cleaned_data.csv',index=False)
```

```python
movies_df[['synopsis','genre','episodes','score']] = movies_df[['synopsis','genre','episodes','score']].astype(str)
movies_df['tags'] = movies_df[['synopsis','genre','episodes','score']].apply(lambda row: " ".join(row), axis=1)
```

```python
movies_df = movies_df.drop(columns = ['synopsis','genre','episodes','score'])
movies_df.head()
```

```python
movies_df.to_csv('data_tags.csv',index=False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies_df['tags'])
cosine_sim = cosine_similarity(tfidf_matrix)
```

```python
def recommend_movies(movie_title):
    index = movies_df[movies_df['title'] == movie_title].index[0]
    dist = list(enumerate(cosine_sim[index]))
    dist.sort(reverse = True,key = lambda x: x[1])
    for i in dist[1:6]:
        print(movies_df.iloc[i[0]]['title'])
```

```python
anime_name = input("Enter anime name: ")
print(f"\nRecommended anime after watching {anime_name}: ")
recommend_movies(anime_name)
```

```python
np.save('cosine_sim.npy', cosine_sim)
```
