```python
import pandas as pd
import numpy as np
```

```python
movies_df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-04-23/tidy_anime.csv')
```

```python
movies_df.head()
```

```python
movies_df.info()
```

```python
movies_df = movies_df[['animeID','name','type','producers','genre','synopsis']]
```

```python
movies_df.head()
```

```python
movies_df['genre'] = movies_df['genre'].astype(str)
grouped_genres = movies_df.groupby('animeID')['genre'].apply(lambda x: ", ".join(x)).reset_index()
```

```python
movies_df = movies_df.merge(grouped_genres, on = 'animeID')
movies_df.head()
```

```python
movies_df = movies_df.drop('genre_x', axis = 1)
movies_df = movies_df.drop_duplicates()
movies_df.head()
```

```python
movies_df['producers'] = movies_df['producers'].astype(str)
grouped_producers = movies_df.groupby('animeID')['producers'].apply(lambda x: ", ".join(x)).reset_index()
```

```python
movies_df = movies_df.merge(grouped_producers, on = 'animeID')
movies_df.head()
```

```python
movies_df = movies_df.drop('producers_x', axis = 1)
movies_df = movies_df.drop_duplicates().reset_index()
movies_df.head()
```

```python
movies_df.isnull().sum()
```

```python
movies_df = movies_df.dropna()
movies_df = movies_df.drop('index', axis = 1)
```

```python
movies_df.to_csv('cleaned_data.csv',index=False)
```

```python
movies_df[['type','synopsis','genre_y','producers_y']] = movies_df[['type','synopsis','genre_y','producers_y']].astype(str)
movies_df['tags'] = movies_df[['type','synopsis','genre_y','producers_y']].apply(lambda row: " ".join(row), axis=1)
```

```python
movies_df = movies_df.drop(columns = ['type','synopsis','genre_y','producers_y'])
movies_df.head()
```

```python
movies_df.to_csv('data_tags.csv',index=False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies_df['tags'])
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def recommend_movies(movie_title):
    index = movies_df[movies_df['name'] == movie_title].index[0]
    dist = list(enumerate(cosine_sim[index]))
    dist.sort(reverse = True,key = lambda x: x[1])
    for i in dist[1:6]:
        print(movies_df.iloc[i[0]]['name'])
```

```python
anime_name = input("Enter anime name: ")
print(f"\nRecommended anime after watching {anime_name}: ")
recommend_movies(anime_name)
```

```python
np.save('cosine_sim.npy', cosine_sim)
```
